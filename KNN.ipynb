```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.neighbors import kneighbors_graph
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
dataset=load_iris()
X_train, X_test, y_train, y_test = train_test_split(dataset["data"],dataset["target"],random_state=13)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X = dataset.data
y = dataset.target
k_values = range(1, 21)
accuracy_scores = []
k_max = k_values[0]
accuracy_max = 0

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = knn.score(X_test, y_test)
    accuracy_scores.append(accuracy)

    if accuracy > accuracy_max:
      accuracy_max = accuracy
      k_max = k

print(f'Max value of k = {k_max} Accuracy = {round(accuracy_max*100, 2)}%')

plt.figure()
plt.bar(k_values, accuracy_scores)
plt.xlabel('k Value')
plt.xticks(k_values)
plt.ylabel('Accuracy Score')
plt.yticks([i/100 for i in range(0,101,5)])
plt.title('KNN Accuracy Score vs. k Value')
plt.show()
print(accuracy_scores)
```
